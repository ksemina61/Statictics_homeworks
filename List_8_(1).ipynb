```python
import numpy as np
from scipy.stats import pearsonr

# Дано
# таблица частот
data = np.array([[1, 0, 1, 0],
                 [2, 4, 4, 2],
                 [0, 1, 0, 1]])

# Значения X и Y
X_values = np.array([2, 3, 4, 5])
Y_values = np.array([3, 4, 5])

# Создаем массивы для X и Y с учетом частот (для правильного вычисления коэффициента корреляции необходимо образовать выборку значений, где каждое значение будет повторяться в соответствии с его частотой)
X = []
Y = []

for i, y in enumerate(Y_values): # В циклах мы заполняем массивы с учетом их частоты
  for j, x in enumerate(X_values):
      count = data[i, j]
      X.extend([x] * count)
      Y.extend([y] * count)

X = np.array(X)
Y = np.array(Y)

# Коэффициент корреляции Пирсона
r, p_value = pearsonr(X, Y)
print(f'Коэффициент корреляции Пирсона: {r}')
print(f'p-value: {p_value}')
```

Это для собственного понимания:

**Коэффициент корреляции Пирсона** показывает линейную зависимость между переменными. Значения близкие к 1 или -1 указывают на сильную зависимость, а значения около 0 — на отсутствие зависимости.

Интерпретация результатов:

Чем ближе значение к 1, тем выше положительная корреляция между показателями.
Если значение ближе к -1, то корреляция отрицательная.
Близкое к 0 значение, включая сам ноль, говорит, что корреляции нет.

**Коэффициент корреляции Спирмена** оценивает монотонную зависимость между переменными и подходит для нелинейных зависимостей.

Интерпретация коэффициента корреляции Спирмена заключается в оценке тесноты связи между двумя рядами сопоставляемых количественных показателей.

Интерпретации значений коэффициента:

p = 1 — идеальная положительная монотонная связь, ранги двух переменных полностью совпадают.
p = -1 — идеальная отрицательная монотонная связь, ранги двух переменных полностью противоположны.
p = 0 — отсутствие монотонной связи, между рангами двух переменных нет никакой связи.


```python
# Во-первых, нужно импортировать библиотеки для вычисления коэффициентов Пирсона и Спирмена
import numpy as np
from scipy.stats import pearsonr, spearmanr

# Дано:
data_a = np.array([(0.7, 0), (1.1, 4), (1.2, 2)])
data_b = np.array([(1, 1), (2, 10), (3, 100)])

# Деление выборки на X и Y
x_a, y_a = data_a[:, 0], data_a[:, 1]
x_b, y_b = data_b[:, 0], data_b[:, 1]

# Коэффициенты корреляции Пирсона и Спирмена для первой выборки (a)
pearson_corr_a, p_value_pearson_a = pearsonr(x_a, y_a)
spearman_corr_a, p_value_spearman_a = spearmanr(x_a, y_a)

# Коэффициенты корреляции Пирсона и Спирмена для второй выборки (b)
pearson_corr_b, p_value_pearson_b = pearsonr(x_b, y_b)
spearman_corr_b, p_value_spearman_b = spearmanr(x_b, y_b)

# Ответ а)
print(f'a) Коэффициент Пирсона: {pearson_corr_a}, p-value: {p_value_pearson_a}, \nКоэффициент Спирмена: {spearman_corr_a}, p-value: {p_value_spearman_a}')
if p_value_pearson_a < 0.05:
    print('Линейная зависимость статистически значима для (a)')
else:
    print('Линейная зависимость незначима для (a)')

if p_value_spearman_a < 0.05:
    print('Монотонная зависимость статистически значима для (a)')
else:
    print('Монотонная зависимость незначима для (a)')

# Ответ b)
print(f'\nb) Коэффициент Пирсона: {pearson_corr_b}, p-value: {p_value_pearson_b}, \nКоэффициент Спирмена: {spearman_corr_b}, p-value: {p_value_spearman_b}')
if p_value_pearson_b < 0.05:
    print('Линейная зависимость статистически значима для (b)')
else:
    print('Линейная зависимость незначима для (b)')

if p_value_spearman_b < 0.05:
    print('Монотонная зависимость статистически значима для (b)')
else:
    print('Монотонная зависимость незначима для (b)')
```

**Теория:**

Коэффициент корреляции  r  принимает значения от -1 до 1:

• Если  r = 1 , это означает идеальную положительную линейную зависимость.

• Если  r = -1 , это означает идеальную отрицательную линейную зависимость.

• Если  r = 0 , это означает отсутствие линейной зависимости.

r = 0.3165522970501863 показывает слабую положительную линейную зависимость между x и y. Имеется тенденция на рост положительной линеной регрессии.

```python
from scipy.stats import pearsonr
# Дано
X = np.array([10, 20, 25, 28, 30])
Y = np.array([5, 8, 7, 12, 14])
# Количество наблюдений
n = len(X)
# Вычисление суммы
sum_X = np.sum(X)
sum_Y = np.sum(Y)
sum_XY = np.sum(X * Y)
sum_X2 = np.sum(X ** 2)
sum_Y2 = np.sum(Y ** 2)
# Коэффициент корреляции Пирсона
numerator = n * sum_XY - sum_X * sum_Y
denominator = np.sqrt((n * sum_X2 - sum_X ** 2) * (n * sum_Y2 - sum_Y ** 2))
r = numerator / denominator
print(f'Коэффициент корреляции Пирсона: {r}')
```

```python
# Потом поняла, что мне понадобится p_value, вычислила отдельно
X = np.array([10, 20, 25, 28, 30])
Y = np.array([5, 8, 7, 12, 14])
# коэффициента корреляции Пирсона и p-значения
r, p_value = pearsonr(X, Y)
print(f'Коэффициент корреляции Пирсона: {r}')
print(f'p_value: {p_value}')
```

```python
if abs(r) > 0.7:
  print('Существует сильная линейная зависимость.')
elif abs(r) > 0.5:
  print('Существует умеренная линейная зависимость.')
else:
  print('Линейная зависимость отсутствует или очень слабая.')

alpha = 0.05
if p_value < alpha:
  print('Линейная зависимость статистически значима.')
else:
  print('Линейная зависимость незначима.')
```
