# **Задача 1.**

 Пусть дана выборка {𝑥1, . . . , 𝑥𝑛} и известно что это результаты независимых
экспериментов, а соответствующая случайная величина 𝜉 распределена с параметрами
E[𝜉] = 𝜇 и D[𝜉] = 𝑑. Проверьте несмещённость и состоятельность оценки  
для 𝜇: 𝜇̂1(𝑥1, . . . , 𝑥𝑛) = 0.21𝑥1 + 0.38𝑥2 + 0.7𝑥3

Дано:

E[𝜉] = 𝜇

D[𝜉] = 𝑑

Найти:

E[𝜇] (несмещенность)?

D[𝜇] --> 0?


Теория:

1. Проверка на несмещенность

Выборочная характеристика, используемая в качестве приближенного значения генеральной характеристики и получаемая вычислением одного числа (точки),
называется точечной **статистической оценкой.**

Решение:

Оценка считается несмещенной, если её математическое ожидание равно истинному значению параметра, то есть E[^μ₁] = μ.
Следовательно, составлю выражение:

 E[μ1] = E[0.21x1 + 0.38x2 + 0.7x3]

 Исходя из свойств математического ожидания E(X + Y) = E(X) + E(Y), где X и Y — случайные величины с конечным математическим ожиданием, применю это свойство в задаче:

 E[μ1] = 0.21E[x1] + 0.38E[x2] + 0.7E[x3]

 Все x[i] являются независимыми и имеют одинаковое распределение с E[x[i]] = μ для всех i:

E[μ1] = 0.21μ + 0.38μ + 0.7μ

E[μ1] = (0.21 + 0.38 + 0.7)μ = 1.29μ

E[^μ₁] != μ (не равно)

Ответ: Следовательно, оценка μ1 является смещенной.

2. Проверка на состоятельность оценки

Теория:

Исправленная выборочная дисперсия S2 есть несмещенная и состоятельная оценка дисперсии D(X) генеральной совокупности.

То есть оценка считается состоятельной, если при увеличении размера выборки n она сходится по вероятности к истинному значению параметра μ.

Чтобы проверить, что оценка состоятельная, нужно убедиться, что она асимптотически несмещённая и её дисперсия стремится к нулю.

Состоятельность означает, что с ростом размера выборки всё менее вероятны значимые отклонения оценки от истинного значения параметра.

**Закон больших чисел в теории вероятностей** — принцип, описывающий результат выполнения одного и того же эксперимента много раз. Согласно закону, среднее значение конечной выборки из фиксированного распределения близко к математическому ожиданию этого распределения. Другими словами, чем больше объём выборки или чем чаще проводятся измерения какого-либо параметра, тем выше вероятность, что результаты окажутся близки к ожидаемым.

Решение:

μ1(x1, x2, x3) = 0.21x1 + 0.38x2 + 0.7x3
Зде
При n --> ∞ средние значения x[i] будут сходиться к μ. В данном случае оценка является состоятельной.

То есть здесь нужно посчитать дисперсию, при увеличении чисел n числитель будет сходиться к мат.ожиданию, будет выполнено условие состоятельности. Сделаю это через программу.

Ответ: Оценка  является состоятельной.

```python
import numpy as np
from scipy import stats
from scipy.stats import chisquare
coeff = np.array([0.21, 0.38, 0.7])
u = 1
d = 1

# Проверка состоятельности
dispersia = np.sum(coeff ** 2) * d

print(f'Дисперсия оценки: {dispersia}')

if dispersia > 0:
    print('Оценка состоятельная, так как дисперсия конечна и при увеличении объёма выборки будет уменьшаться.')
else:
    print('Оценка несостоятельная, так как дисперсия не убывает.')
```

Пункт b)

Гипотезы:

H0 =  дети из данной группы семей владеют языком так же хорошо, как в общем дети их возраста.

H1 =  дети из данной группы семей владеют языком лучше, чем в общем дети их возраста ( 1.7 > 1.4).

Опять нужно найти среднее значение:

$ x = \frac{1,8 + 1,9 + 1,2 + 2,7 + 2,3}{5} = \frac{9,9}{5} = 1,98$

$ M = \frac{(1,8 - 1,98)^2 + (1,9 - 1,98)^2 + (1,2 - 1,98)^2 + (2,7 - 1,98)^2 + (2,3 - 1,98)^2}{n-1} = \frac{(-0,18)^2 + (-0,08)^2 + (-0,78)^2 + (0,72)^2 + (0,32)^2}{4} = \frac{0,0324 + 0,0064 + 0,6084 + 0,5184 + 0,1024}{4} = \frac{1,268}{4} = 0,317$

𝜎 = √(M) = √0,317 = 0,563



$ t-student = \frac{x - μ}{𝜎/√n} = \frac{1,98 - 1,4}{0,563/√5} = \frac{0,58}{0,251} = 2,31 $

Здесь мы используем двусторонний тест.
Затем также, как и в первом случае определю критические значения:

Смотрю таблицу для одностороннего теста.

t(0,10, 4) = 1,533

Для alpha = 0,10: 2,31 > 1,533 - отвергаю H0

Вывод: На уровне значимости 10% есть основания отвергнуть H0, и на уровне значимости 5% нет оснований отвергать H0.

```python
import numpy as np
import scipy.stats as stats
children_data = np.array([1.8, 1.9, 1.2, 2.7, 2.3])
mu = 1.4
n = len(children_data)
mean = np.mean(children_data)
std = np.std(children_data, ddof=1)

alpha_10 = 0.10
alpha_5 = 0.05

# t-статистика
t_stat = (mean - mu) / (std / np.sqrt(n))

# Критические значения t-распределения для одностороннего теста
t_crit_10 = stats.t.ppf(1 - alpha_10, df=n-1)
t_crit_5 = stats.t.ppf(1 - alpha_5, df=n-1)

print(f'Среднее по выборке: {mean}')
print(f'Стандартное отклонение по выборке: {std}')
print(f't-статистика: {t_stat}')
print(f'Критическое значение t для уровня значимости 10%: {t_crit_10}')
print(f'Критическое значение t для уровня значимости 5%: {t_crit_5}')

# Проверка гипотез
if t_stat > t_crit_10:
    print('Отклоняем нулевую гипотезу на уровне значимости 10%')
else:
    print('Не хватает оснований для отклонения нулевой гипотезы на уровне значимости 10%')

if t_stat > t_crit_5:
    print('Отклоняем нулевую гипотезу на уровне значимости 5%')
else:
    print('Не хватает оснований для отклонения нулевой гипотезы на уровне значимости 5%')
```
