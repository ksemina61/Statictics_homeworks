```python
import math
from scipy.integrate import quad
# Функция с формулой нормального распределения
def normal_distribution(x, M, sigma):
  return (1 / (math.sqrt(2 * math.pi) * sigma) * math.exp(-((x- M)**2)/(2 * sigma**2)))
# Кумулятивная функция распределения (cdf) — это функция, которая даёт вероятность того, что случайная величина примет значение, меньшее или равное заданному.
def normal_cdf(x, M, sigma):
  # Нужно проинтегрировать эту функцию
  def take_integral(s):
    return normal_distribution(s, M, sigma)
  return quad(take_integral, -math.inf, x)[0]
# [0] - Функция quad возвращает кортеж с результатами - первым элементом является значение интеграла, а вторым элементом является оценка погрешности интегрирования. Используя [0], берется только первое значение (результат интеграла)
# Параметры нормального распределения
M = 1
sigma = 2
# Пункт а) P(X <= 2)
p_a = normal_cdf(2, M, sigma)
print(f'P(X <= 2) = {p_a}')
# Пункт b) P(X <= -1)
p_b = normal_cdf(-1, M, sigma)
print(f'P(X <= -1) = {p_b}')
# Пункт с) P(X <= - 10)
p_c = normal_cdf(-10, M, sigma)
print(f'P(X <= -10) = {p_c}')
# Пункт d) P(-1 < X <= 2)
p_d = normal_cdf(2, M, sigma) - normal_cdf(-1, M, sigma)
print(f'P(-1 < X <= 2) = {p_d}')
# Пункт e) P(2 < X <= 3)
p_e = normal_cdf(3, M, sigma) - normal_cdf(2, M, sigma)
print(f'P(2 < X <= 3) = {p_e}')
# Пункт f) P(X > 2)
# Для этой вероятности P(X > a) нужно вычесть 1, т.к сумма вероятностей = 1.
p_f = 1 - normal_cdf(2, M, sigma)
print(f'P(X > 2) = {p_f}')
# Пункт g) P(X > -1)
p_g = 1 - normal_cdf(-1, M, sigma)
print(f'P(X > -1) = {p_g}')
```

```python
import math
from scipy.integrate import quad
def normal_distribution(x, M, sigma):
  return (1 / (math.sqrt(2 * math.pi) * sigma) * math.exp(-((x- M)**2)/(2 * sigma**2)))
def normal_cdf(x, M, sigma):
  def take_integral(z):
    return normal_distribution(z, M, sigma)
  return quad(take_integral, -math.inf, x)[0]
# Дано (параметры распределения Y):
M_y = 3
sigma_y = 2

# (c) P(Y <= 2)
p_c = normal_cdf(2, M_y, sigma_y)
print("P(Y <= 2) =", p_c)
# (d) P(Y <= -10)
p_d = normal_cdf(-10, M_y, sigma_y)
print("P(Y <= -10) =", p_d)
# (e) P(-1 < Y <= 2)
p_e = normal_cdf(2, M_y, sigma_y) - normal_cdf(-1, M_y, sigma_y)
print("P(-1 < Y <= 2) =", p_e)
# (f) P(Y > 2)
p_f = 1 - normal_cdf(2, M_y, sigma_y)
print("P(Y > 2) =", p_f)
```
